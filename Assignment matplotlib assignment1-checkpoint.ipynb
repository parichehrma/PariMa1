```python
x=df.Age
y=df.Survived
plt.hist(x, density=True, color='blue', bins=np.arange(0, 100), alpha=.5) 
plt.hist(y, density=True, color='red', bins=np.arange(0, 100), alpha=.5)
plt.title('Normed histograms')
plt.xlabel('Age')
plt.show()
```

```python
##### In the above figure,the blue figure shows passenger's age distribution and the red figure shows survival rate.
```
