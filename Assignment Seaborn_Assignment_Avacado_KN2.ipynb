```python
# Choose two continuous variables, and plot them three different ways.
# I choose avocados PLU 4046 and 4225 sold as two continuous variables and displays in three different ways.
# First displays with histplot:
plt.hist(avocado['4046'], color='red', alpha=0.5, label='4046', bins=5)
plt.hist(avocado['4225'], color='blue', alpha=0.5, label='4225', bins=5)
plt.xlabel('Distribution of avocados sold')
plt.legend(loc='upper right')
plt.title('Plot 1: Compare Total number of avocados PLU 4046 and 4225 sold')
plt.show()
```

```python
# Second displays with lineplot:
avocado.plot(kind='line', x='4046', y='4225', title='Plot 2:Compare Total number of avocados PLU 4046 and 4225 sold') 
plt.show() 
```

Shows distribution of 4046 and 4225  avocados sold.

```python
# Third displays with scatterplot:
sns.set(style='white')
g = sns.lmplot(x='4046', y='4225', data=avocado, fit_reg=False, scatter_kws={'alpha':0.5}, height=6) 
g.set_xlabels("4046")
g.set_ylabels("4225")
plt.title('Plot 3: Scatterplot: Compare Total number of avocados PLU 4046 and 4225 sold')
plt.show()
g = sns.lmplot(x='4046', y='4225', data=avocado, fit_reg=True, scatter_kws={'alpha':0.5}, height=6)
g.set_xlabels("4046")
g.set_ylabels("4225")
plt.title('Plot 3:Scatterplot with regression line: Compare Total number of avocados PLU 4046 and 4225 sold')
plt.show()
```

Scatterplot  shows more clear information. It shows more clear distribution of avocados sold in both and they have direct relationship which means if avocado's PLU 4046 sold more the avocado's PLU 4225 sold more too.

```python
# Choose one continuous variable and one categorical variable, and plot them six different ways.
# I choose Total Volume as a continuous variable and type as a categorical variable.
# First displays with barhplot:
avocado.groupby('type')['Total Volume'].mean().plot(kind='barh', color='blue')
plt.xlabel('type')
plt.ylabel('Total Volume')
plt.show()
```

The barhplot shows the higher value of total number of avocados sold is conventional type.

```python
# Second displays with scatterplot:
ax=plt.scatter(x='Total Volume', y='type', data= avocado)
plt.title('Plot 2:Distribution of Total Number of Avocados Sold by type')
plt.xlabel="Total Volume"
plt.ylabel="Type"
plt.show()
```

The scatterplot shows the higher value of total number of avocados sold is conventional type.

```python
# Third displays with strip pointplot:
sns.set(style='whitegrid')
ax=sns.pointplot(x='Total Volume', y='type', data= avocado, kind='strip', dodge=True, join=False, ci=95)
plt.title('Plot 3:Distribution of Total Number of Avocados Sold by type')
ax.set(xlabel="", ylabel='Type')
sns.despine(offset=10, trim=True)
plt.show()
```

The stripplot shows the higher value of total number of avocados sold is conventional type.

```python
# Forth displays with barplot:
sns.set(style='darkgrid')
ax=sns.catplot(x='Total Volume', y='type', data= avocado, kind="bar")
plt.title('Plot 4:Distribution of Total Number of Avocados Sold by type')
ax.set(xlabel="", ylabel='Type')
sns.despine(offset=10, trim=True)
plt.show()
```

The bar catplot shows the higher value of total number of avocados sold is conventional type.

```python
# Fifth displays with violinplot:
sns.violinplot(x="type", y="Total Volume", data=avocado)
plt.title('Plot 5:Distribution of Total Number of Avocados Sold by type')
plt.show()
```

```python
# I choose AveragePrice,Total Volume,avocado PLU 4046,avocado PLU 4225 as a continuous variables,type as a categorical variable.
x = avocado[avocado.type=='organic']['AveragePrice']
y = avocado[avocado.type!='organic']['AveragePrice']
plt.hist(x, bins=6, alpha=0.5, label='organic')
plt.hist(y, bins=6, alpha=0.5, label='conventional')
plt.title('Distribution of AveragePrice by type')
plt.legend(loc='upper right')
plt.show()
```

The distribution of the average price of a single avocado for organic type varies more than the conventional type.

```python
x = avocado[avocado.type=='organic']['Total Volume']
y = avocado[avocado.type!='organic']['Total Volume']
plt.title('Distribution of Total Volume by type')
plt.hist(x, bins=6, label='organic')
plt.hist(y, bins=6, label='conventional')
plt.legend(loc='upper right')
plt.show()
```

Total number of avocados sold is conventional type.

```python
x = avocado[avocado.type=='organic']['4046']
y = avocado[avocado.type!='organic']['4046']
plt.title('Distribution of Avocado PLU 4046 Sold by type')
plt.hist(x, bins=6, label='organic')
plt.hist(y, bins=6, label='conventional')
plt.legend(loc='upper right')
plt.show()
```

Total number of avocados with PLU 4046 sold is conventional type.

```python
x = avocado[avocado.type=='organic']['4225']
y = avocado[avocado.type!='organic']['4225']
plt.title('Distribution of Avocado PLU 4225 Sold by type')
plt.hist(x, bins=6, label='organic')
plt.hist(y, bins=6, label='conventional')
plt.legend(loc='upper right')
plt.show()
```

Total number of avocados with PLU 4225 sold is conventional type.

The distribution of the average price of a single avocado for organic type varies more than the conventional type but total number of avocados sold in conventional type is more.
