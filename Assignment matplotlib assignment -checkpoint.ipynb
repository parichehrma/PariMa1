```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# df= pd.read_excel(r'C:\Users\mebra.DESKTOP-L12LJA6\Thinkful Data Science Projects\titanic.xlsx')
# #print(df)
df= pd.read_csv('titanic.csv')
#print(df)
```

```python
df.groupby('Survived').PassengerId.count()
```

```python
df.groupby('Survived').PassengerId.count().plot(kind='bar', color='blue', x='Survived', y='PassengerId')                                             
plt.title('Survival Rate')
plt.show()
```

```python
##### The above figure shows survival passenger rate is less.
```

```python
df.groupby('Survived').Age.mean()
```

```python
df.groupby('Survived').Age.mean().plot(kind='barh', color="green")
plt.title('Survival Average Age Rate')
plt.show()
```

```python
##### The above figure shows average age of survived people is less than 30.
```

```python
df.groupby('Sex').Survived.sum()
```

```python
df.groupby('Sex').Survived.sum().plot(kind='pie', title='Survival Rate by Sex')
labels = [r'Organge Male', r'Blue Female']  
plt.legend(labels, loc="best") 
plt.show()
```

```python
##### The above figure shows female Survival rate is more than male.
```

```python
df.groupby(['Pclass','Sex']).Survived.sum()
```

```python
df.groupby(['Pclass','Sex']).Survived.sum().plot(kind='bar', color='green')
```

```python
##### The above figure shows distribution of male and female in different class.
```

```python
df.groupby('Age').Survived.sum().plot(kind='line', color='red', title='Survival Age Range')
plt.show 
```

```python
##### The above figure shows survival age range of 20 to 40 years old are more than other age ranges.
```

```python
df.plot(kind='scatter',x='Pclass',y='Age',color='red')
plt.show()
```

```python
# The figure above shows age range in different class.
```

```python
df.groupby('Pclass').Survived.sum().plot(kind='pie')
labels = [ r'Blue Class 1', r'Organge Class 2', r'Green Class 3',]  
plt.legend(labels, loc="best") 
plt.show()
    
```

```python
##### The above figure shows survival class range which shows passengers in class 1 survived more than others.
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['Age'], color='purple')
plt.ylabel('Age')
plt.title('Age Rate')

plt.subplot(1, 2, 2)
plt.scatter(x = df['Sex'], y = df['Age'], color='green')
plt.ylabel('Survived')
plt.title('Survived Rate')

plt.tight_layout()
plt.show()
```

```python
x=df.Age
y=df.Survived
plt.hist(x, density=True, color='blue', bins=np.arange(0, 100), alpha=.5) 
plt.hist(y, density=True, color='red', bins=np.arange(0, 100), alpha=.5)
plt.title('Normed histograms')
plt.xlabel('Age')
plt.show()
```

```python
##### In the above figure,the blue figure shows passenger's age distribution and the red figure shows survival rate.
```
